```python
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model
```

```python
# Define the input and output arrays
n1 = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])
n2 = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])
```

```python

# Convert arrays to TensorFlow tensors and reshape them
n = tf.constant(n1, dtype=tf.float32)
m = tf.constant(n2, dtype=tf.float32)

n = tf.reshape(n, [-1, 1])  # Reshape to column vector
m = tf.reshape(m, [-1, 1])  # Reshape to column vector
```

```python
# Create a sequential model
model = tf.keras.Sequential([
   tf.keras.layers.Dense(100,activation=None),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001),
              loss=tf.keras.losses.mae,
              metrics=["mae"])

# Fit the model
model.fit(n,m,epochs=500)#jitnay epoch ziada utna ghat loss
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation=None,input_shape=[1],name="Input_layer"),
    tf.keras.layers.Dense(1,name="Output_layer"),
    tf.keras.layers.Dense(1),
])
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(lr=0.001),
    metrics=["mae"]
)
model.fit(X_train,y_train,epochs=280)
```

```python
def plot_models(train_data=X_train,
                lable_train=y_train,
                test_data=X_test,
                test_label=y_test,
                predictions=y_pred):
    plt.figure(figsize=(10,7))
    plt.scatter(train_data,lable_train,c='r',label="Training Data")
    plt.scatter(test_data,test_label,c='g',label="Testing data")
    plt.scatter(test_data,predictions,c='b',label="Comparing Predicted values")
    plt.legend()
```
